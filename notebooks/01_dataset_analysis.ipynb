```python
from pathlib import Path

import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATASET_PATH = Path("../data/raw/PlantVillage")
```

```python
print(DATASET_PATH)
```

```python
if DATASET_PATH.exists():
    print("✅ Dataset found")
else:
    print("❌ Dataset not found")
```

```python
folders = [
    folder.name
    for folder in DATASET_PATH.iterdir()
    if folder.is_dir()
]

folders
```

```python
classes = sorted(
    [
        folder.name
        for folder in DATASET_PATH.iterdir()
        if folder.is_dir()
    ]
)

print("Total Classes :", len(classes))
```

```python
for cls in classes:
    print(cls)
```

```python
total_images = 0

for cls in classes:
    folder = DATASET_PATH / cls

    total_images += len(list(folder.glob("*")))

print("Total Images :", total_images)
```

```python
crops = set()

for cls in classes:
    crop = cls.split("___")[0]
    crops.add(crop)

print("Crop Species :", len(crops))
```
