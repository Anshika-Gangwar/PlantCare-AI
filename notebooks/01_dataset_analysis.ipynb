```python
from pathlib import Path

import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
folders = [
    folder.name
    for folder in DATASET_PATH.iterdir()
    if folder.is_dir()
]

folders
```

```python
classes = sorted(
    [
        folder.name
        for folder in DATASET_PATH.iterdir()
        if folder.is_dir()
    ]
)

print("Total Classes :", len(classes))
```

```python
for cls in classes:
    print(cls)
```

```python
total_images = 0

for cls in classes:
    folder = DATASET_PATH / cls

    total_images += len(list(folder.glob("*")))

print("Total Images :", total_images)
```

```python
crops = set()

for cls in classes:
    crop = cls.split("___")[0]
    crops.add(crop)

print("Crop Species :", len(crops))
```

```python
image_sizes = {}
```

```python
for cls in classes:

    folder = DATASET_PATH / cls

    for img_path in folder.glob("*"):

        img = Image.open(img_path)

        size = img.size

        if size in image_sizes:
            image_sizes[size] += 1
        else:
            image_sizes[size] = 1
```

```python
print(image_sizes)
```

```python
for size, count in image_sizes.items():

    print(f"Image Size : {size} --> {count} images")
```

```python
most_common_size = max(image_sizes, key=image_sizes.get)

print("Most Common Image Size:", most_common_size)
```

```python
#Class Distribution Analysis

class_counts = {}

for cls in classes:
    folder = DATASET_PATH / cls
    class_counts[cls] = len(list(folder.glob("*")))

class_counts
```

```python
df_classes = pd.DataFrame(
    class_counts.items(),
    columns=["Class", "Image Count"]
)

df_classes
```

```python
df_classes = df_classes.sort_values(
    by="Image Count",
    ascending=False
)

df_classes
```

```python
plt.figure(figsize=(15,8))

plt.bar(
    df_classes["Class"],
    df_classes["Image Count"]
)

plt.xticks(rotation=90)

plt.title("Number of Images per Class")

plt.xlabel("Plant Disease Class")

plt.ylabel("Number of Images")

plt.tight_layout()

plt.show()
```

```python
largest = df_classes.iloc[0]

smallest = df_classes.iloc[-1]

print("Largest Class")
print(largest)

print()

print("Smallest Class")
print(smallest)
```

```python
average_images = df_classes["Image Count"].mean()

print(f"Average Images per Class: {average_images:.2f}")
```

```python
std = df_classes["Image Count"].std()

print(f"Standard Deviation : {std:.2f}")
```

```python
median = df_classes["Image Count"].median()

print(f"Median Images : {median}")
```

```python
#Image Format Analysis

formats = {}

for cls in classes:

    folder = DATASET_PATH / cls

    for img_path in folder.glob("*"):

        suffix = img_path.suffix.lower()

        formats[suffix] = formats.get(suffix,0)+1

formats
```

```python
different_formats = []

for cls in classes:
    folder = DATASET_PATH / cls

    for img_path in folder.glob("*"):
        if img_path.suffix.lower() != ".jpg":
            different_formats.append(img_path)

for file in different_formats:
    print(file)
```

```python
for file in different_formats:
    img = Image.open(file)

    print(f"File: {file.name}")
    print(f"Format: {img.format}")
    print(f"Size: {img.size}")
    print(f"Mode: {img.mode}")
    print("-" * 40)
```

```python
#Corrupted Image Detection

corrupted = []

for cls in classes:

    folder = DATASET_PATH / cls

    for img_path in folder.glob("*"):

        try:
            img = Image.open(img_path)
            img.verify()

        except Exception:

            corrupted.append(img_path)
```
